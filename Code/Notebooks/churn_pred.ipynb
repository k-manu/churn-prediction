```python
df.describe()
```

```python
print("\nChurn distribution:\n", df['Exited'].value_counts(normalize=True))
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm",cbar=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
numeric_features = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
for feature in numeric_features:
    plt.figure(figsize=(8, 4))
    sns.kdeplot(data=df, x=feature, hue="Exited", fill=True)
    plt.title(f"{feature} Distribution by Churn")
    plt.show()
```

```python
categorical_features = ['Geography', 'Gender', 'HasCrCard', 'IsActiveMember']
for feature in categorical_features:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=feature, hue="Exited")
    plt.title(f"{feature} vs Churn")
    plt.show()
```
